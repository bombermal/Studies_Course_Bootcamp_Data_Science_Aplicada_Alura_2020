Imports


```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

```python
# Load file
url = "https://github.com/alura-cursos/agendamento-hospitalar/raw/main/dados/A160324189_28_143_208.csv"
dataRaw = pd.read_csv(url, encoding="latin-1", sep=";", skiprows=3, skipfooter = 12, thousands=".", decimal=",",engine='python')

# Options
pd.options.display.float_format="{:.4f}".format
# Show data
dataRaw
```

```python
# Remove number from frist column names
dataRaw['Unidade da Federação'] = dataRaw['Unidade da Federação'].apply(lambda x: x[3:])
data = dataRaw.set_index("Unidade da Federação")
# Replace invalid values
data = data.replace('-', 0).astype(float)
data
```

## Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada

```python
#Seleciono todas as colunas exceto 'Total'
cols = data.columns.tolist()[:-1]

#gráfico
fig, ax = plt.subplots(figsize=(12,8))
data[cols].T.plot(kind="line", figsize=(15,5), ax=ax)
ax.legend(loc='upper left', ncol = 2, fancybox=True, shadow=True)
ax.yaxis.set_major_formatter(tck.StrMethodFormatter("{x:,.2f}"))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.title("Valor processamento por ano/mês segundo UF")
plt.show()
```

## Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
#Sample de gráficos
fig, ax = plt.subplots(figsize=(12,8))
data.loc[:,"2008/Fev":"2020/Ago"].sample(5).T.plot(kind="line", ax=ax)
plt.legend(bbox_to_anchor=(1.18,1), loc="upper right", fancybox=True, shadow=True)
ax.yaxis.set_major_formatter(tck.StrMethodFormatter("{x:,.2f}"))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.title("Valor processamento por ano/mês segundo UF")
plt.show()
```
